```
import pypowsybl as pp
```

```
net = pp.network.load("cgmes.zip")
```

```
glsk_file = "glsk_file.xml"
crac_file = "crac_file.xml"

rao_runner = pp.rao.create_rao()
rao_runner.set_glsk_file_source(net, glsk_file)
rao_runner.set_crac_file_source(net, crac_file)
rao_parameters=pp.rao.Parameters()

rao_runner.run(net, rao_parameters)
```

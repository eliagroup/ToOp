```
import json
import os
import re

import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```
folder = "./jax/run_greedy_t4"
```

```
filename_regex = r"^bsdf_(?P<bsdf>\d*)_inj_(?P<inj>\d*)\.json$"

results = []

for file in os.listdir(folder):
    match = re.match(filename_regex, file)
    if match:
        bsdf = int(match.group("bsdf"))
        inj = int(match.group("inj"))
        with open(os.path.join(folder, file), "r") as f:
            contents = json.load(f)
            results.append(
                {
                    "bsdf": bsdf,
                    "inj": inj,
                    "mean": np.mean(contents["times"]),
                    "std": np.std(contents["times"]),
                }
            )
    else:
        continue

df = pd.DataFrame(results)
df.index = pd.MultiIndex.from_frame(df[["bsdf", "inj"]])
```

```
# Loadflows per second peak
12065760 / df["mean"].min()
```
